```python
### using inbuilt functions
import numpy as np
n=int(input("Enter the number of observation:"))
mylist_=[]
for i in range(n):
    a=int(input("enter the "+str(i+1) + "st element:"))
    mylist_.append(a)
myarray=np.array(mylist_)

#calculating mean

mean_1=np.mean(myarray)
print("mean:",mean_1)

#calculating median

median_1=np.median(myarray)
print("median:",median_1)

#calculating standard deviation

std_dev=np.std(myarray)
print("standard deviation:",std_dev)

#calculating variance

varian_ce=np.var(myarray)
print("variance",varian_ce)

#calculating range

min_1=np.min(myarray)
max_1=np.max(myarray)
range_1=max_1-min_1
print("range:",range_1)

#calculating mode
import statistics as st

mo_de=st.mode(myarray)
print("mode:",mo_de)

# #mode using scipy
# from scipy import stats
# mode2=stats.mode(myarray)
# print("The mode is",mode2)
```
