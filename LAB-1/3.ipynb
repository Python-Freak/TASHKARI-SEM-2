```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x + 4
plt.title("Linear")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x, y, color='m', linestyle='--', marker='.')
```

```python
x = np.arange(0,10, 0.1)
y = np.sin(x)
plt.title("Sin (X)")
plt.xlabel("X - Axis")
plt.ylabel("Sin (x)")
plt.plot(x,y, linestyle='-', marker='.', color='r')
```

```python
plt.subplot(2,1,1)
plt.title("Sin X")
plt.plot(x,y)
plt.subplot(2,1,2)
plt.title("Cos X")
plt.plot(x, np.cos(x))
```

```python
x = np.arange(0, 3*np.pi, 0.01)
sin = np.sin(x)
cos = np.cos(x)
plt.subplot(2,2,1)
plt.plot(x, sin)
plt.title("Sine")
plt.subplot(2,2,2)
plt.plot(x, cos)
plt.title("Cosine")
plt.subplot(2,2,3)
plt.plot(x, cos)
plt.title("Cosine")
plt.subplot(2,2,4)
plt.plot(x, sin)
plt.title("Sine")

plt.subplots_adjust(hspace=0.5, wspace = 0.5)
plt.suptitle("Trigo")
```

```python
y = np.sin(x)
z = np.cos(x)
plt.title("Sin vs Cosine")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x, y, color='r', label="Sin")
plt.plot(x, z, color='b', label="Cos")
plt.xticks(np.arange(0,3*np.pi, 0.5))
plt.yticks(np.arange(-1.2,1.2, 0.1))
```

```python
s1 = [19, 16, 20, 19]
s2 = [13, 10, 16, 13]
sub = ['MATH', "PHYSICS", "ENVIRONMENTAL", "ELECTRICAL"]
plt.scatter(np.arange(0, 4), s1, label="Student - 1", color='b', marker='^')
plt.scatter(np.arange(0, 4), s2, label="Student - 1", color='r', marker='o')
plt.xticks(np.arange(0,4), sub, rotation=30)
plt.yticks(np.arange(0,22, 1))
```

```python
x = [5, 8, 10]
y = [12, 16, 6]
x2 = [6, 9, 11]
y2 = [6, 15, 7]

plt.bar(x, y, color = 'b', align = 'center')
plt.bar(x2, y2, color = 'g', align = 'center')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins = [0,20,40,60,80,100], color = 'b', edgecolor='k', linewidth = 0.5)
plt.title("histogram")
plt.show()
```
